```python
dist_list = list(collections.Counter(dfr["PdDistrict"]).keys())
dist_list = dict(map(lambda x :(x,dist_list.index(x)),dist_list))
dist_list
```

学習用のテーブルを用意する

```python
train_data = pd.DataFrame(columns = ['Year','Month','Date','Period','DayOfWeek','PdDistrict','Address'])
train_data
```

```python
(lambda x: x+1)(2)
```

```python
train_data["Year"] = dfr["Dates"].apply(lambda x: x.split(" ")[0].split("-")[0])
train_data["Month"] = dfr["Dates"].apply(lambda x: x.split(" ")[0].split("-")[1])
train_data["Date"] = dfr["Dates"].apply(lambda x: x.split(" ")[0].split("-")[2])
train_data["Period"] = dfr["Dates"].apply(lambda x: int(x.split(" ")[1].split(":")[0])//3)
train_data["DayOfWeek"] = dfr["DayOfWeek"].apply(lambda x: week_list[x])
train_data["PdDistrict"] = dfr["PdDistrict"].apply(lambda x: dist_list[x])
train_data["Address"] = dfr["Address"].apply(lambda x: address_list[x])
```

```python
train_data_y = []
for x in dfr["Category"]:
    train_data_y.append(category_list.index(x))
```

```python
train_data_y[:5]
```

```python
train_data = train_data.values
```

学習

```python
from sklearn import tree
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(train_data, train_data_y)
```

```python
from sklearn.externals.six import StringIO
import os
```

```python
with open("result.dot", 'w') as f:
    f = tree.export_graphviz(clf, out_file=f)
```

```python
predicted = clf.predict(train_data)
predicted
```

```python
sum(predicted == train_data_y) / len(train_data_y)
```

```python
test_data = pd.DataFrame(columns = ['Year','Month','Date','Period','DayOfWeek','PdDistrict','Address'])
test_data["Year"] = dfr_test["Dates"].apply(lambda x: x.split(" ")[0].split("-")[0])
test_data["Month"] = dfr_test["Dates"].apply(lambda x: x.split(" ")[0].split("-")[1])
test_data["Date"] = dfr_test["Dates"].apply(lambda x: x.split(" ")[0].split("-")[2])
test_data["Period"] = dfr_test["Dates"].apply(lambda x: int(x.split(" ")[1].split(":")[0])//3)
test_data["DayOfWeek"] = dfr_test["DayOfWeek"].apply(lambda x: week_list[x])
test_data["PdDistrict"] = dfr_test["PdDistrict"].apply(lambda x: dist_list[x])
test_data["Address"] = dfr_test["Address"].apply(lambda x: address_list[x])
```

```python
test_data[:3]
```

```python
predicted = clf.predict(test_data)
predicted
```

```python
test_result = pd.DataFrame(np.zeros((len(dfr_test),len(category_list))))
test_result.columns=category_list
for i, x in enumerate(predicted):
    test_result.ix[i,x] = 1
```

```python
test_result[category_list] = test_result[category_list].astype(int)
```
