```python
shot_type_list = list(collections.Counter(dfr["combined_shot_type"]).keys())
shot_type_dict = dict(zip(shot_type_list,range(0,len(shot_type_list))))
combined_shot_type_list = [shot_type_dict[shot_type] for shot_type in dfr["combined_shot_type"]]
```

```python
shot_zone_list = list(collections.Counter(dfr["shot_zone_area"]).keys())
shot_zone_dict = dict(zip(shot_zone_list,range(0,len(shot_zone_list))))
shot_zone_area_list = [shot_zone_dict[shot_zone] for shot_zone in dfr["shot_zone_area"]]
print(shot_zone_dict)
```

Prepare datasets

```python
datasets = pd.DataFrame({'loc_x_list': loc_x_list,
                           'loc_y_list': loc_y_list,
                           'remaining_time_list': remaining_time_list,
                           'playoffs': dfr["playoffs"],
                           'shot_made_flag': dfr["shot_made_flag"],
                           'combined_shot_type': combined_shot_type_list,
                           'shot_id':dfr["shot_id"],
                           #'shot_zone_area': shot_zone_area_list,
                          })
datasets
```

make train_data,test_data

```python
datasets = datasets.astype(str)
train_data = datasets[datasets["shot_made_flag"] != "nan"]
train_data_y = train_data["shot_made_flag"]
train_data_x = train_data.drop(["shot_made_flag"],axis=1)
train_data_x = train_data_x.drop(["shot_id"],axis=1)
```

```python
test_data = datasets[datasets["shot_made_flag"] == "nan"]
test_data = test_data.drop("shot_made_flag",axis=1)
shot_id = test_data["shot_id"].values
test_data = test_data.drop("shot_id",axis=1)
```

```python
train_data_x = train_data_x.values
train_data_y = train_data_y.values
test_data = test_data.values
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
y_pred = gnb.fit(train_data_x, train_data_y).predict(train_data_x)
#np.set_printoptions(threshold=np.inf)
y_pred
```

```python
print((train_data_y == y_pred).sum()/(train_data_x.shape[0])*100)
```

```python
y_pred = gnb.fit(train_data_x, train_data_y).predict(test_data)
```

```python
y_pred = pd.DataFrame({"shot_made_flag": y_pred})
shot_id = pd.DataFrame({"shot_id": shot_id})
```

```python
pd.concat([shot_id,y_pred],axis=1).to_csv('pred_result.csv',index=False)
```

ここから考察
シュート地点における影響から考えてみる

```python
remaining_time = (60*dfr["minutes_remaining"]+dfr["seconds_remaining"])
shotframe = pd.DataFrame({"shot_made_flag": dfr["shot_made_flag"],
                        "loc_x":dfr["loc_x"],
                        "loc_y":dfr["loc_y"],
                        "remaining_time":remaining_time ,
                        "combined_shot_type":combined_shot_type_list, 
                        "shot_zone_area": shot_zone_area_list,  
                        })
sinkshot = shotframe[shotframe["shot_made_flag"] == 1.0]
not_sinkshot = shotframe[shotframe["shot_made_flag"] == 0.0]
```

```python
plt.scatter(x=sinkshot["loc_x"][:300],y=sinkshot["loc_y"][:300],c="red",alpha=0.5)
plt.scatter(x=not_sinkshot["loc_x"][:300],y=not_sinkshot["loc_y"][:300],c="blue",alpha=0.5)
```

距離が極端に遠い場合においては外す割合の方が高いことがわかるが、中間距離以下だと影響は見受けられない。
そのため、遠い地点を判定する上ではシュート地点は学習において有用なラベルと言える。（と思う）

```python
plt.hist([sinkshot["remaining_time"],not_sinkshot["remaining_time"]],normed=True,bins=20,color=["red","blue"],alpha=1.0)
```

残り時間が0に近い時とゲーム開始後は外す確率の方が高いことがわかる。
しかし、それ以外の時間帯は若干数ではあるが決める確率の方が高いことが見て取れる。
そのため、残り時間は学習をするにあたって有用であると考えることができる（と思う）

```python
plt.hist([sinkshot["combined_shot_type"],not_sinkshot["combined_shot_type"]],color=["red","blue"],alpha=1.0)
```

```python
shot_type_dict
```

シュートのタイプは大きく影響が見られる。Jump Shootの場合は外す割合が高いがLayupとDunkに関しては決める割合が高いことが見て取れる。

```python
plt.hist([sinkshot["shot_zone_area"],not_sinkshot["shot_zone_area"]],color=["red","blue"],alpha=1.0)
```

```python
shot_zone_dict
```

シュートエリアはCenter(C)つまり真ん中付近以外は外す割合の方が高くなることが見て取れた。

以上より、今回使用した各ラベルは有用性が期待出来るものばかりだったので、残念な結果になった理由としては、識別方法(ナイーブベイズ)が適切なものでなかったか、学習に使用していないラベルの中に学習に高い効果をもたらすラベルがあったなどその他の理由が考えられる。

考察は終わったが、loc_xとloc_yの値は個別に見るのではなくて相互関係がある気がしたので試してみる。

ゴールからのユークリッド距離をとる

```python
loc_list = (np.hypot(dfr["loc_x"],dfr["loc_y"])//100).astype(int)
loc_list.unique()
```

```python
datasets = pd.DataFrame({'loc':loc_list,
                           'remaining_time_list': remaining_time_list,
                           'playoffs': dfr["playoffs"],
                           'shot_made_flag': dfr["shot_made_flag"],
                           'combined_shot_type': combined_shot_type_list,
                           'shot_id':dfr["shot_id"],
                           'shot_zone_area': shot_zone_area_list,
                          })
```

```python
datasets = datasets.astype(str)
train_data = datasets[datasets["shot_made_flag"] != "nan"]
train_data_y = train_data["shot_made_flag"]
train_data_x = train_data.drop(["shot_made_flag"],axis=1)
train_data_x = train_data_x.drop(["shot_id"],axis=1)
test_data = datasets[datasets["shot_made_flag"] == "nan"]
test_data = test_data.drop("shot_made_flag",axis=1)
shot_id = test_data["shot_id"].values
test_data = test_data.drop("shot_id",axis=1)
train_data_x = train_data_x.values
train_data_y = train_data_y.values
test_data = test_data.values
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
y_pred = gnb.fit(train_data_x, train_data_y).predict(train_data_x)
#np.set_printoptions(threshold=np.inf)
y_pred
print((train_data_y == y_pred).sum()/(train_data_x.shape[0])*100)
```

```python
y_pred = gnb.fit(train_data_x, train_data_y).predict(test_data)
y_pred = pd.DataFrame({"shot_made_flag": y_pred})
shot_id = pd.DataFrame({"shot_id": shot_id})
pd.concat([shot_id,y_pred],axis=1).to_csv('pred_result2.csv',index=False)
```
